```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline

print ("All loaded")
```

```python
df["Revenue"] = df["Revenue"].str.replace(',','')
df["Revenue"] = df["Revenue"].str.replace('$','')
df['Revenue'] = pd.to_numeric(df['Revenue'],errors='coerce')
```

```python
df.head(10)
```

```python
df["Expenses"] = df["Expenses"].str.replace('Dollars','')
df["Expenses"] = df["Expenses"].str.replace(',','')
df['Expenses'] = pd.to_numeric(df['Expenses'],errors='coerce')
```

```python
df.head(10)
```

```python
df["Growth"] = df["Growth"].str.replace('%','')
df['Growth'] = pd.to_numeric(df['Growth'],errors='coerce')
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Industry'].fillna(method='pad', inplace=True)
```

```python
#df['State'].fillna(method='pad', inplace=True)
mapping = dict(df.dropna()[['City', 'State']].values)
df['State'] = df['State'].fillna(df['City'].map(mapping))
```

```python
df['State'].fillna(method='pad', inplace=True)
```

```python
df.head(15)
```

```python
df.iloc[[*range(83, 87), *range(265, 269)]]
```

```python
df['Inception'].fillna(df['Inception'].median(), inplace=True)
```

```python
#df['Employees'].fillna(df['Employees'].mean(), inplace=True)
df['Employees'].fillna(df['Employees'].median(), inplace=True)
```

```python
# df['Revenue'] = df['Revenue'].fillna(df['Revenue'].mean()) 
#calculate mean value Revenue by industry for construction and implement for corporation
std_dict = df[['Revenue', 'Industry']].groupby(['Industry']).mean().to_dict()['Revenue']
df['Revenue'] = df.Revenue.fillna(df.Industry.map(std_dict))
```

```python
df.head(10) # use filter to investigate which value are Nan that look
```

```python
#df['Expenses'] = df['Expenses'].fillna(df['Expenses'].std()) 
#calculate std value by industry for construction and implement for corporation
std_dict1 = df[['Expenses', 'Industry']].groupby(['Industry']).std().to_dict()['Expenses']
df['Expenses'] = df.Expenses.fillna(df.Industry.map(std_dict1))
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Profit'] = df['Profit'].fillna(df['Revenue'] - df['Expenses'])
```

```python
df.head(10)
```

```python
#df['Growth'] = df['Growth'].fillna(df['Growth'].std())
std_dict2 = df[['Growth', 'Industry']].groupby(['Industry']).std().to_dict()['Growth']
df['Growth'] = df.Growth.fillna(df.Industry.map(std_dict2))
```
